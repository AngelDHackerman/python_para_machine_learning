Series de tiempo

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.dates import DateFormatter
```

```python
dates = pd.date_range(start= '2023-01-01', periods=100)
values = np.random.rand(100).cumsum()  # cumsum: La suma acumulativa significa que cada elemento en el nuevo array es la suma del elemento actual y todos los anteriores en la secuencia.
data = pd.DataFrame({'Date': dates, 'Value': values})
print(data)
```

```python
# Crear grafico de lineas
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data['Date'], data['Value'], color='green')
plt.xticks(rotation=45)
plt.title('Serie de tiempo con formato en las fechas')
plt.xlabel('Date')
plt.xlabel('Value')

plt.show()
```

```python
# Generar fechas mensuales y valores acumulativos
dates = pd.date_range(start= '2023-01-01', periods=12, freq='ME') # 'M' indica frecuencia mensual
values = np.random.uniform(50, 100, size=12) # (media, desviacion_estandar)
data = pd.DataFrame({'Date': dates, 'Value': values})
print(data)
```

```python
# Crear grafico de lineas
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data['Date'], data['Value'], color='green', marker='o') # Agregar marcador para cada mes
plt.xlabel('Fecha')
plt.ylabel('Valor')

# Establecer el limite inferior del eje y para iniciar en 40
ax.set_ylim(40, data['Value'].max() + 10) # Ajustar para dar un margen extra en el valor mximo

# formatear el eje x para mostrar solo el mes y el ao
date_form = DateFormatter("%b %Y") # Formato: "Mes Ao (por ejemplo, "Jan 2023")
ax.xaxis.set_major_formatter(date_form)
ax.set_xticks(data['Date'])
plt.xticks(rotation=45, fontsize=8)

plt.show()
```
