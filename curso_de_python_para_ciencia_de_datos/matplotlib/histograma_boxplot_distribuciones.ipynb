Graficos de histograma y boxplot para distribuciones

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(170, 10, 200) # media, desviacion estandar, N. de datos
print(data)
```

```python
plt.hist(data, color='salmon', bins=20, edgecolor='black', alpha=0.6)
plt.title('Distribucion de alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')
plt.show()
```

```python
np.random.seed(0)
ages = [np.random.normal(30, 5, 100),
        np.random.normal(35, 5, 100),
        np.random.normal(40, 5, 100)]
print(ages)
```

```python
plt.boxplot(ages, patch_artist=True, notch=True, vert=True,
            labels=['Group 1', 'Grupo 2', 'Grupo 3'])

plt.title('Distribucion de edades')
plt.ylabel('Edades')
plt.show()
```
