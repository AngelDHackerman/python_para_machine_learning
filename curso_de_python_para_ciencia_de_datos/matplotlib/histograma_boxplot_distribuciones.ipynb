Graficos de histograma y boxplot para distribuciones

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(170, 10, 200) # media, desviacion estandar, N. de datos
print(data)
```

```python
plt.hist(data, color='salmon', bins=20, edgecolor='black', alpha=0.6)
plt.title('Distribucion de alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')
plt.show()
```
