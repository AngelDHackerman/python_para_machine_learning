```python
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

hours = [2, 3, 4, 5, 6, 7, 8]
exam_scores_studen_1 = [55, 60, 65, 70, 75, 80, 85]
exam_scores_studen_2 = [50, 58, 63, 69, 74, 78, 83]

# Crear grafico de dispersion de dos estudiantes
plt.scatter(hours, exam_scores_studen_1, marker='o', color='green', linestyle='-', label='Estudiante 1')
plt.scatter(hours, exam_scores_studen_2, marker='s', color='red', linestyle='--', label='Estudiante 2')

plt.title('Relación entre horas estudiadas y el puntaje de 2 estudiantes')
plt.xlabel('Horas')
plt.ylabel('Puntaje')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

hours = [2, 3, 4, 5, 6, 7, 8]
exam_scores_studen_1 = [55, 60, 65, 70, 75, 80, 85]
exam_scores_studen_2 = [50, 58, 63, 69, 74, 78, 83]

# Crear grafico de dispersion de dos estudiantes
plt.plot(hours, exam_scores_studen_1, marker='o', color='green', linestyle='-', linewidth= 2, label='Estudiante 1')
plt.plot(hours, exam_scores_studen_2, marker='s', color='red', linestyle='--', linewidth= 2, label='Estudiante 2')

plt.title('Relación entre horas estudiadas y el puntaje de 2 estudiantes')
plt.xlabel('Horas')
plt.ylabel('Puntaje')

plt.show()
```
