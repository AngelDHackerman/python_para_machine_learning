```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definimos los vectores u y v como arrays de NumPy
u = np.array([1, 2, 3])
v = np.array([3, 6, 9])

# Calculamos la suma de los vectores
u_plus_v = u + v

# Mostramos los resultados de los cálculos
print("Vector u:", u)
print("Vector v:", v)
print("Suma de u y v (u + v):", u_plus_v)

# Configuramos la visualización en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # 1 fila, 1 columna, 1 subplot = 111

# Dibujamos el vector u en color azul
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='blue', label="Vector u", arrow_length_ratio=0.25) # 0,0,0 Estos tres primeros argumentos representan el punto de origen del vector, u[0], u[1], u[2] Estos tres valores representan las componentes del vector en los ejes X,Y,𝑍, respectivamente.

# Dibujamos el vector v en color rojo
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='red', label="Vector v", arrow_length_ratio=0.1)

# Dibujamos el vector u + v en color verde
ax.quiver(0, 0, 0, u_plus_v[0], u_plus_v[1], u_plus_v[2], color='green', label="Vector u + v", arrow_length_ratio=0.1)

# Ajustes adicionales para el gráfico
ax.set_xlim([0, 5])
ax.set_ylim([0, 10])
ax.set_zlim([0, 15])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

# Mostramos el gráfico
plt.show()

# Como puedes observar, el vector verde representa la combinación de 𝑢 y 𝑣, comenzando desde el origen y extendiéndose hasta el punto resultante de la suma.
```
